matplotlib tutorial

```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

t = np.arange(0,12,0.01)
y = np.sin(t)
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(t, y)
plt.plot(t, np.cos(t))
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sine and cosine wave')
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(t, y, label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sine and cosine wave')
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.plot(t, y, lw=3, label='sin')
plt.plot(t, np.cos(t), 'green', label='cos')
plt.legend()
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sine and cosine wave')
plt.show()
```

```
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.show()
```

```
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.scatter(t,y)
plt.show()
```

```
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]
plt.figure(figsize=(10,6))
plt.plot(t,y, linestyle = 'dashed')
plt.show()
```

```
t = np.array([0,1,2,3,4,5,6])
y = np.array([1,4,5,8,9,5,3])
plt.figure(figsize=(10,6))
plt.plot(t,y, linestyle = 'dashed', marker='o', markerfacecolor='green', markersize=10)
plt.xlim([-0.5, 6.5])
plt.ylim([0.5, 9.5])
plt.show()
```

```
t = np.array([0,1,2,3,4,5,6])
y = np.array([1,4,5,8,9,5,3])
colormap = y
plt.figure(figsize=(10,6))
plt.scatter(t,y, marker='>', s = 50, c = colormap)
plt.colorbar()
plt.xlim([-0.5, 6.5])
plt.ylim([0.5, 9.5])
plt.show()
```

```
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)

plt.figure(figsize=(10,6))
plt.plot(s1, label = 's1')
plt.plot(s2, label = 's2')
plt.plot(s3, label = 's3')
plt.legend()
plt.show()
```

```
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```

# 네이버 영화 평점!

```python
import pandas as pd
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200128"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
soup
```

```python
for m in soup.find_all('div', 'tit5'):
    print(m.a.string)
```

```python
date = pd.date_range('2019-1-1', periods=365, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td','point')[n].string for n in range(0, end)])
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie
```

```python
import numpy as np

score = []
for point in movie['point']:
    score.append(float(point))
movie['score'] = score
movie_unique = pd.pivot_table(movie, index=['name'],values='score', aggfunc=np.mean)

movie_best = movie_unique.sort_values(by='score', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == ["보헤미안 랩소디"]')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.plot(tmp['date'], tmp['score'])
plt.legend(loc='best')
plt.grid()
plt.show()
```

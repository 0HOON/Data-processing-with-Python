# 네이버 영화 평점!

```python
import pandas as pd
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200128"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")
soup
```

```python
for m in soup.find_all('div', 'tit5'):
    print(m.a.string)
```

```python
date = pd.date_range('2019-1-1', periods=365, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, "html.parser")
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td','point')[n].string for n in range(0, end)])
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie
```

```python
import numpy as np

score = []
for point in movie['point']:
    score.append(float(point))
movie['score'] = score
movie_unique = pd.pivot_table(movie, index=['name'],values='score', aggfunc=np.mean)

movie_best = movie_unique.sort_values(by='score', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == ["보헤미안 랩소디"]')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.plot(tmp['date'], tmp['score'])
plt.legend(loc='best')
plt.grid()
plt.show()
```

- 영화별 날짜에 따른 별점 변화

# 시카고 샌드위치 맛집 소개

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = 'http://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')

soup
```

```python
soup.find_all('div', 'sammy')
```

```python
tmp_one = soup.find_all('div', 'sammy')[0]
```

```python
tmp_one.find(class_='sammyRank').text
```

```python
tmp_one.find(class_='sammyListing').text
```

```python
tmp_one.find('a')['href']
```

```python
import re

tmp_string = tmp_one.find(class_='sammyListing').text

re.split(('\n|\r\n'), tmp_string)

print(re.split(('\n|\r\n'), tmp_string)[0])
print(re.split(('\n|\r\n'), tmp_string)[1])
```

```python
from urllib.parse import urljoin

rank = []
main_menu = []
cafe_name = []
url_add = []

url_base = 'http://www.chicagomag.com'

list_soup = soup.find_all('div', 'sammy')

for item in list_soup:
    rank.append(item.find(class_='sammyRank').text)
    
    tmp_string = item.find(class_='sammyListing').text
    
    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0])
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href']))
```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add)
```

```python
data = {'Rank': rank, 'Menu': main_menu, 'Cafe': cafe_name, 'URL': url_add}
df = pd.DataFrame(data, columns=['Rank', 'Cafe', 'Menu', 'URL'])
df.head()
```

```python
df.to_csv('./data/03. best_sandwiches_list_chicago.csv', sep=',', encoding='UTF-8') 
```

```python
html = urlopen(df['URL'][0])
soup_tmp = BeautifulSoup(html, 'html.parser')
soup_tmp.find('p', 'addy').text
```

```python
price_tmp = soup_tmp.find('p', 'addy').text
price_tmp.split()
```

```python
from tqdm import tqdm_notebook

price = []
address = []
for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'html.parser')
    gettings = soup_tmp.find('p', 'addy').text.split()
    price.append(gettings[0][:-1])
    address.append(' '.join(gettings[1:-2]))
```

```python
df['Price'] = price
df['Address'] = address

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace=True)
df.head()
```

```python
df.to_csv('./data/03. best_sandwiches_list_chicago2.csv', sep=',', encoding='UTF-8')
```

## 지도에 나타내기

```python
import folium
import googlemaps
import pandas as pd
import numpy as np

df = pd.read_csv('./data/03. best_sandwiches_list_chicago2.csv', index_col=0)
gmaps_key = "AIzaSyDR4bwV1Qywzy2nsZoUSCij1KQvCH_tcGg"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n] + ', ' + 'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)         
```

```python
df['lat'] = lat
df['lng'] = lng
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping)
mapping
```

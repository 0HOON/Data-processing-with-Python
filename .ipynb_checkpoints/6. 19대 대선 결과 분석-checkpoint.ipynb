```python
sido_list = driver.find_element_by_xpath("""//*[@id="cityCode"]""").find_elements_by_tag_name("option")
sido_names_values = [option.text for option in sido_list]
sido_names_values
```

```python
sido_names_values
```

```python
from tqdm import tqdm_notebook

sido_name_list = []
sigun_name_list = []

for sido_value in tqdm_notebook(sido_names_values):
    element = driver.find_element_by_id("cityCode")
    element.send_keys(sido_value)
    
    time.sleep(1)
    
    sigun = driver.find_element_by_xpath("""//*[@id="townCode"]""")
    sigun_list = sigun.find_elements_by_tag_name("option")
    
    sigun_names_values = [option.text for option in sigun_list]
    sigun_names_values = sigun_names_values[1:]
    
    for sigun_value in sigun_names_values:
        sido_name_list.append(sido_value)
        sigun_name_list.append(sigun_value)
```

```python
election_result = pd.DataFrame({"광역시도":sido_name_list, "시군":sigun_name_list})

election_result.head()
```

```python
from bs4 import BeautifulSoup 

def get_vote_info(n):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    tmp = soup.find_all('td', 'alignR')
    tmp_values = [float(tmp_val.get_text().replace(',', '')) for tmp_val in tmp[1:5]]
    
    pop[n] = tmp_values[0]
    moon[n] = tmp_values[1]
    hong[n] = tmp_values[2]
    ahn[n] = tmp_values[3]
```

```python
def fail_procedure(n):
    pop[n] = np.nan
    moon[n] = np.nan
    hong[n] = np.nan
    ahn[n] = np.nan    
```

```python
pop = [np.nan]*len(election_result)
moon = [np.nan]*len(election_result)
hong = [np.nan]*len(election_result)
ahn = [np.nan]*len(election_result)

len(pop), len(moon), len(hong), len(ahn)
```

```python
for n in tqdm_notebook(election_result.index):
    try:
        element = driver.find_element_by_id("cityCode")
        element.send_keys(election_result['광역시도'][n])
        
        time.sleep(0.5)
        
        element = driver.find_element_by_id("townCode")
        element.send_keys(election_result['시군'][n])
        
        driver.find_element_by_xpath("""//*[@id="searchBtn"]""").click()
        
        time.sleep(0.5)
        
        get_vote_info(n)
        
        time.sleep(0.5)
        
    except:
        print("-----Error-----")
        fail_procedure(n)
```

```python
election_result['pop'] = pop
election_result['moon']= moon
election_result['hong'] = hong
election_result['ahn'] = ahn
```

```python
def cut_char_sigu(name):
    return name if len(name) == 2 else name[:-1]
```

```python
import re

sigun_candi = ['']*len(election_result)

for n in election_result.index:
    each = election_result['시군'][n]
    if each[:2] in ['수원', '성남', '안양', '안산', '고양', '용인', '청주', '천안', '전주', '포항', '창원']:
        sigun_candi[n] = re.split('시', each)[0] + ' ' + cut_char_sigu(re.split('시', each)[1])
    
    else:
        sigun_candi[n] = cut_char_sigu(each)
        
sigun_candi
```

```python
ID_candi = [sido_candi[n] + ' ' + sigun_candi[n] for n in range(0, len(sigun_candi))]

ID_candi = [name[1:] if name[0] == ' ' else name for name in ID_candi]
ID_candi = [name[:2] if name[:2] == '세종' else name for name in ID_candi]

ID_candi
```

## 데이터 시각화 하기 - 5장의 함수 이용

```python
BORDER_LINES = [
    [(5, 1), (5,2), (7,2), (7,3), (11,3), (11,0)], # 인천
    [(5,4), (5,5), (2,5), (2,7), (4,7), (4,9), (7,9), 
     (7,7), (9,7), (9,5), (10,5), (10,4), (5,4)], # 서울
    [(1,7), (1,8), (3,8), (3,10), (10,10), (10,7), 
     (12,7), (12,6), (11,6), (11,5), (12, 5), (12,4), 
     (11,4), (11,3)], # 경기도
    [(8,10), (8,11), (6,11), (6,12)], # 강원도
    [(12,5), (13,5), (13,4), (14,4), (14,5), (15,5), 
     (15,4), (16,4), (16,2)], # 충청북도
    [(16,4), (17,4), (17,5), (16,5), (16,6), (19,6), 
     (19,5), (20,5), (20,4), (21,4), (21,3), (19,3), (19,1)], # 전라북도
    [(13,5), (13,6), (16,6)], # 대전시
    [(13,5), (14,5)], #세종시
    [(21,2), (21,3), (22,3), (22,4), (24,4), (24,2), (21,2)], #광주
    [(20,5), (21,5), (21,6), (23,6)], #전라남도
    [(10,8), (12,8), (12,9), (14,9), (14,8), (16,8), (16,6)], #충청북도
    [(14,9), (14,11), (14,12), (13,12), (13,13)], #경상북도
    [(15,8), (17,8), (17,10), (16,10), (16,11), (14,11)], #대구
    [(17,9), (18,9), (18,8), (19,8), (19,9), (20,9), (20,10), (21,10)], #부산
    [(16,11), (16,13)], #울산
#     [(9,14), (9,15)], 
    [(27,5), (27,6), (25,6)],
]


def drawKorea_2(targetData, blockedMap, cmapname):
    # for data including minus values
    gamma = 0.75
    
    whitelabelmin = 20.
    
    datalabel = targetData
    
    tmp_max = max([ np.abs(min(blockedMap[targetData])), np.abs(max(blockedMap[targetData]))])
    
    vmin = -tmp_max
    vmax = tmp_max
    
    mapdata = blockedMap.pivot_table(index='y', columns='x', values= targetData)
    masked_mapdata = np.ma.masked_where(np.isnan(mapdata), mapdata)
    
    plt.figure(figsize=(9,11))
    plt.pcolor(masked_mapdata, vmin=vmin, vmax=vmax, cmap=cmapname, edgecolor='#aaaaaa', linewidth=0.5)

    for idx, row in blockedMap.iterrows():
        if len(row['ID'].split()) == 2:
            dispname = '{}\n{}'.format(row['ID'].split()[0], row['ID'].split()[1])
        elif row['ID'][:2] == '고성':
            dispname = '고성'
        else:
            dispname = row['ID']

        if len(dispname.splitlines()[-1]) >= 3:
            fontsize, linespacing = 9.5, 1.5
        else:
            fontsize, linespacing = 11, 1.2

        annocolor = 'white' if row[targetData] > whitelabelmin else 'black'
        plt.annotate(dispname, (row['x'] + 0.5, row['y'] + 0.5), weight='bold', fontsize=fontsize, \
                     ha='center', va='center', linespacing=linespacing, color=annocolor)

    for path in BORDER_LINES:
        ys, xs = zip(*path)
        plt.plot(xs, ys, c='black', lw=1.5)

    plt.gca().invert_yaxis()

    plt.axis('off')

    plt.tight_layout()
    
    cb = plt.colorbar(shrink=.1, aspect=10)
    cb.set_label(datalabel)
    
    plt.show()
```

```python
drawKorea_2('moon_vs_hong', final_elect_data, 'RdBu')
```

```python
drawKorea_2('moon_vs_ahn', final_elect_data, 'RdBu')
```

```python
drawKorea_2('ahn_vs_hong', final_elect_data, 'RdBu')
```

```python
import folium
import json
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
pop_folium = final_elect_data.set_index('ID')

del pop_folium['광역시도']
del pop_folium['시군']

pop_folium.head()
```

```python
geo_path = 'data/05. skorea_municipalities_geo_simple.json'
geo_data = json.load(open(geo_path, encoding='utf-8'))

def showMap(targetData, cmapname):
    map = folium.Map(location=[36.2002, 127.054], zoom_start=7)
    map.choropleth(geo_data=geo_data, data = pop_folium[targetData], columns = [pop_folium.index, pop_folium[targetData]],\
                  fill_color=cmapname, key_on='feature.id')
    return map
    
showMap('moon_vs_hong', 'RdBu')
```
